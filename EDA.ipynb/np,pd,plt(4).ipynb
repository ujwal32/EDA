```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
a = np.array([[[1,2,3], [4,5,6]]])
b = np.array([[np.random.randint(1,10,6)]])
c = np.array([np.random.randint(1,10,6)])
```

```python
b
```

```python

a, b, c = (a.ravel(), b.ravel(), c.ravel())
```

```python
b
```

```python
dataframe_data = {'length':a, 'breadth':b, 'width':c}
```

```python
df=pd.DataFrame(data=dataframe_data)
```

```python
df
```

```python
df.plot(x="length", y="breadth", kind="line", title="Testing line graph")
plt.show()
```

```python
df.plot(x="length", y="breadth", kind="bar", title="Testing line graph")
plt.savefig("figure_2.jpg", dpi=200)
plt.show()
```
