## Make the similar scatter plot using pyplot and download the tips dataset from internet and load it using pandas

```python
import matplotlib
from matplotlib import pyplot as plt #alis
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.set_theme(style="darkgrid")
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", palette="viridis")
plt.title("Bar Plot Example")
plt.show(
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="sex")
plt.title("IR class scatter plot")
plt.show()
```
